```python
import tkinter as tk

def button_click(number):
    global val
    val += str(number)
    data.set(val)

root = tk.Tk()
root.title("Calculator")

val = ""
data = tk.StringVar()


entry = tk.Entry(root, textvariable=data, font=('Arial', 20), justify='right')
entry.grid(row=0, columnspan=5)


button7 = tk.Button(root, text='7', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(7))
button7.grid(row=1, column=0)

button8 = tk.Button(root, text='8', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(8))
button8.grid(row=1, column=1)

button9 = tk.Button(root, text='9', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(9))
button9.grid(row=1, column=2)

buttonplus = tk.Button(root, text='+', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click('+'))
buttonplus.grid(row=1, column=3)

button4 = tk.Button(root, text='4', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(4))
button4.grid(row=2, column=0)

button5 = tk.Button(root, text='5', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(5))
button5.grid(row=2, column=1)

button6 = tk.Button(root, text='6', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(6))
button6.grid(row=2, column=2)

buttonminus = tk.Button(root, text='-', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click('-'))
buttonminus.grid(row=2, column=3)

button1 = tk.Button(root, text='1', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(1))
button1.grid(row=3, column=0)

button2 = tk.Button(root, text='2', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(2))
button2.grid(row=3, column=1)

button3 = tk.Button(root, text='3', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(3))
button3.grid(row=3, column=2)

buttonmul = tk.Button(root, text='*', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click('*'))
buttonmul.grid(row=3, column=3)

buttonC = tk.Button(root, text='C', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: clear())
buttonC.grid(row=4, column=0)

button0 = tk.Button(root, text='0', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click(0))
button0.grid(row=4, column=1)

buttonequal = tk.Button(root, text='=', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: calculate())
buttonequal.grid(row=4, column=2)

buttondiv = tk.Button(root, text='/', font=("Arial", 12, "bold"), bd=12, height=2, width=6, command=lambda: button_click('/'))
buttondiv.grid(row=4, column=3)

def clear():
    global val
    val = ""
    data.set("")

def calculate():
    global val
    try:
        result = str(eval(val))  # Evaluate the expression
        data.set(result)
        val = result
    except:
        data.set("Error")
        val = ""

root.mainloop()
```
